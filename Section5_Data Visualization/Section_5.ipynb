TO CREATE 2 DIFF PLOTS IN THE SAME CHART EG UNITED AIRLINES (UA) & AMERICAN AIRLINES (AA)

text, mtext, lines, points can only come after plot command of 1 data

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"])
points(arr_delay[carrier=="AA"],dep_delay[carrier=="AA"],pch=3)
```

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"],col="Red")
points(arr_delay[carrier=="AA"],dep_delay[carrier=="AA"],col="Green")
```

HOW TO PUT GRAPHS SIDE BY SIDE OR ON MULTIPLE PANELS


mfrow=multiframerow; c=vector combination; 1=no of rows; 2=no of columns (to have UA and AA side by side); if you had 4 graphs then you could have had 2 rows and2 columns or your wish

```r
par(mfrow=c(1,2))
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="UA")
plot(arr_delay[carrier=="AA"],dep_delay[carrier=="AA"], main="AA")
```

Equal scale for x and y axis.

```r
par(mfrow=c(1,2))
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="UA", 
     xlim=c(0,1000), ylim=c(0,1000))
plot(arr_delay[carrier=="AA"],dep_delay[carrier=="AA"], main="AA",
     xlim=c(0,1000), ylim=c(0,1000))
```

Restoring the frame for a single plot

```r
par(mfrow=c(1,1))
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="UA")
```

Placing of Text on the graph!

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time")
text(x=100, y=400, label="United Airways")
```

Adj means to adj text to left or right or middle of text
0=left
1=right
0.5=middle

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time")
text(x=100, y=300, label="United Airways", adj=1)
```

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time")
text(x=100, y=300, label="United Airways", adj=0.5)
```

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time")
text(x=100, y=300, label="United Airways", adj=0)
```

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time")
text(x=400, y=0, label="Sandeep")
```

mtext is Text outside the margin of the 4 boxes;
Bottom is side 1;
Left is side 2;
Top is side 3;
Right is side 4.

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time")
mtext(text="UA= United Airways", side=3)
```

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time")
mtext(text="inminutes", side=2,adj=1)
```

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time",col="red")
```

CHANGE COLOR OF MAIN TITLE;<BR> The below code shows size of graph (cex), color of graph (col), color of main title (col.main), font of main title (font.main) where font.main=1 (Normal), 2 (Bold), 3 (Italics), 4 (Italics and Bold)

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time",cex=2, col="red", col.main="red", font.main=3)
```

CHANGE OF COLOR AND FONT OF LABELS ON SIDES AND BOTTOM (col.lab; font.lab)

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time",cex=2, col="red", col.main="red", font.main=3, col.lab="green", font.lab=4)
```

ADDITION OF TEXT IN GRAPH IN A DIFF COLOR AND FONT


```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time",cex=2, col="red", col.main="red", font.main=3, col.lab="green", font.lab=4)
text(100,400,"United Airways", col="blue",font=2)
```

HAVING 2 TEXTS IN THE GRAPH

```r
plot(arr_delay[carrier=="UA"],dep_delay[carrier=="UA"], main="Arrival vs Departure time",cex=2, col="red", col.main="red", font.main=3, col.lab="green", font.lab=4)
text(100,400,"United Airways", col="blue",font=2)
text(100,300,"United Airways",col="pink",font=2,cex=2)
```
